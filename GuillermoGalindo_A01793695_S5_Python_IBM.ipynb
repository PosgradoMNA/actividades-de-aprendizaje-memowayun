Lectura de data para hacer ejemplos

```python
path= 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv'
df = pd.read_csv(path,header=None)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
## asignamos el nombre de las columnas
df.columns=headers
```

```python
# QUitar valores NA 
```

```python
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
df.dropna()
```

```python
# para guardar los cambios
df.dropna(subset=['price'], axis=0, inplace=True)
```

```python
# para reemplazar missing values 
```

```python
mean=df['wheel-base'].mean()
```

```python
df['wheel-base'].replace(np.nan, mean)
```

**Data Formatting**

```python
df['city-mpg'] = 235/df['city-mpg']

df.rename(columns={'city-mpg':'city-L/100k'}, inplace=True)
```

```python
# Verificar y asignar el tipo de dato correcto 
```

```python
df['price'].tail()
```

```python
df.dtypes
```

```python
df['price'] = df['price'].astype("double")
```

**Data Normalization**

Escalar las variables en el mismo rango, de forma que sean comparables y utilizables en algunos algorítmos

- Simple Feature Scaling (dividir por el máximo valor)

```python
df['length'] = df['length']/df['length'].max()
```

- Min Max (Restar el mín y dividir entre la diferencia entre el máx y el make

```python
df['length'] = (df['length'] - df['length'].min())/(df['length'].max() - df['length'].min())
```

- Z Score (restar la media y dividir entre la desviación estándar)

```python
df['length'] = (df['length']-df['length'].mean())/df['length'].std()
```

**Binning**

```python
bins = np.linspace(min(df['engine-size']), max(df['engine-size']),4)
```

```python
group_names = ['low','medium','hihg']
```

```python
df['engine-size_binned'] = pd.cut(df['engine-size'],bins, labels=group_names, include_lowest=True)
```
