```python
# Liberarías a usar 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn.distributions import stats
```

```python
path= 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv'
df = pd.read_csv(path,header=None)
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
## asignamos el nombre de las columnas
df.columns=headers
```

```python
df.replace("?", np.nan, inplace = True)
df.replace("None", np.nan, inplace = True)
df.dropna()
```

# Analysis of Variance ANOVA

Correlación de diferentes grupos de variables categóricas con la variables de respuesta u otra variables

Anova: 
- F test score: variación (diferencia) entre la media de dos grupos dividida por la variación dentro del grupo de análisis
- p value: nivel de confianza

- F pequeña poca diferencia entre grupos 

- F grande más diferencia entre las medias de los grupos


Para que exista diferencia entre grupos el F Score debe ser mayor o igual a 1

```python
df_anova = df[['make','price']]
```

```python
grouped_anova = df_anova.groupby(['make'])
```

```python
anova_results = stats.f_oneway(grouped_anova.get_group('honda')['price'], grouped_anova.get_group('subaru')['price'])
```

```python
anova_results
```

El resultado anterior es menor a 1, y sin significancia, no hay diferencia entre grupos

```python
anova_results_1 = stats.f_oneway(grouped_anova.get_group('honda')['price'], grouped_anova.get_group('jaguar')['price'])
```

```python
anova_results_1
```

F score es. muy grande, si hay diferencia entre grupos y es significativa por p value muy pequeño

# Pearson Correlation

Correlation coefficient: 

- Correlation coeff close to +1: large positive relationship
- Correlation coeff close to -1: large negativ relationship
- Correlation coeff close to 0: NO relationship

p value: 
- p-value < 0.001 STRONG certainty in the result 
- p-value < 0.05 Moderate certainty in the result 
- p-value < 0.1 Weak certainty in the result 
- p-value > 0.1 NO certainty in the result 



```python
df['horsepower'] = df['horsepower'].astype("double")
df.replace([np.inf, -np.inf], np.nan)
df.dropna(inplace=True)
pearson_coef, p_value = stats.pearsonr(df['horsepower'], df['price'])
```

```python
pearson_coef
```

```python
p_value
```
